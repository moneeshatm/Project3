```python
import mediapipe as mp
import cv2
```

```python
mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils
```

```python
data=mp_hands.Hands(max_num_hands=1)
```

```python
video = cv2.VideoCapture(0)
tipid = [4, 8, 12, 16, 20]
while True:
    suc, img = video.read()
    img = cv2.flip(img, 1)
    img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    result = data.process(img1)
    lmlist = []
    gesture = ""
    if result.multi_hand_landmarks:
        for handlms in result.multi_hand_landmarks:
            for id, lm in enumerate(handlms.landmark):
                cx, cy = lm.x, lm.y
                lmlist.append([id, cx, cy])
            if len(lmlist) == 21:
                fingercount = []
                # Thumb logic 
                if lmlist[12][1] < lmlist[20][1]:  # Right hand
                    fingercount.append(1 if lmlist[4][1] < lmlist[3][1] else 0)
                else:  # Left hand
                    fingercount.append(1 if lmlist[4][1] > lmlist[3][1] else 0)
                # Other fingers
                for i in range(1, 5):
                    fingercount.append(
                        1 if lmlist[tipid[i]][2] < lmlist[tipid[i] - 2][2] else 0
                    )
                if fingercount == [0, 0, 0, 0, 0]:
                    gesture = "FIST"
                elif fingercount == [1, 1, 1, 1, 1]:
                    gesture = "OPEN PALM"
                elif fingercount == [0, 1, 1, 0, 0]:
                    gesture = "PEACE"
                elif fingercount == [1, 0, 0, 0, 0]:
                    gesture = "THUMBS UP"
                elif fingercount == [0, 1, 0, 0, 0]:
                    gesture = "POINTING"
                elif fingercount == [1, 1, 0, 0, 0]:
                    gesture = "L SHAPE"
                elif fingercount == [1, 0, 0, 0, 1]:
                    gesture = "CALL ME"
                elif fingercount == [0, 1, 1, 1, 0]:
                    gesture = "THREE"
                elif fingercount == [0, 1, 0, 0, 1]:
                    gesture = "ROCK"
                elif fingercount == [1, 1, 0, 0, 1]:
                    gesture = "OK"

                cv2.putText(img, gesture, (40, 400),
                            cv2.FONT_HERSHEY_COMPLEX, 2, (0, 255, 0), 3)

            mp_drawing.draw_landmarks(
                img, handlms, mp_hands.HAND_CONNECTIONS)

    cv2.imshow("Hand Gesture Recognition", img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

video.release()
cv2.destroyAllWindows()
```
